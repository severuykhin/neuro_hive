```python
import time
import random
import base64
import requests

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options

def load_img(src, index, type):
    filename = str(index) + '-' + type + '.jpg'
    directory = './0' + type + '/'
    r = requests.get(src, allow_redirects=True)
    with open(directory + filename, 'wb') as f:
        f.write(r.content)

mobile_emulation = {
    "deviceMetrics": { "width": 360, "height": 640, "pixelRatio": 3.0 },
    "userAgent": "Mozilla/5.0 (Linux; Android 4.2.1; en-us; Nexus 5 Build/JOP40D) AppleWebKit/535.19 (KHTML, like Gecko) Chrome/18.0.1025.166 Mobile Safari/535.19" 
}

chrome_options = Options()

browser = webdriver.chrome.webdriver.WebDriver(executable_path='./chromedriver', chrome_options=chrome_options)

time.sleep(random.randrange(1,2))

browser.get("https://yandex.ru/images/search?text=%D0%BC%D1%83%D0%B6%D1%87%D0%B8%D0%BD%D0%B0%20%D0%B2%D0%BA%20%D1%84%D0%BE%D1%82%D0%BE&from=tabbar")

for i in range(3):
    print('scroll')
    browser.execute_script("window.scrollTo(0, 3000 * " + str(i) + " );")
    time.sleep(5)
```

```python
has_glasses_items = browser.find_elements_by_css_selector('img')
print(len(has_glasses_items))
for i in range(len(has_glasses_items)):
    item = has_glasses_items[i]
    src = item.get_attribute('src')
    if src:
        load_img(src, str(i) + '2','2')
        print(src)
        
print('Finished images loading')
```
